```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# O que ja sabemos?

empresas = ['Apple', 'Google', 'Amazon', 'Microsoft', 'Tencent', 'McDonalds', 'Visa', 'Facebook', 'Alibaba', 'Louis Vuitton']
```

```python
bilhoes= [947.062, 819.573, 705.646, 611.460, 214.023, 196.526, 191.032, 186.421, 169.966, 124.273]
```

```python
# Já é possível plotar, só com as variáveis

plt.plot(empresas, bilhoes)
```

```python
# Adicionando título
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
```

```python
# Adicionando rótulo no eixo X e Y
plt.plot(empresas, bilhoes)
plt.title('Empresas Mais Ricas do Mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
```

```python
# Redimensionando

plt.rcParams["figure.figsize"] = (15, 8)

plt.plot(empresas, bilhoes)
plt.title('Empresas Mais Ricas do Mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
plt.show()
```

```python
# Conhecendo o fontsize

plt.plot(empresas, bilhoes)
plt.title('Empresas Mais Ricas do Mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.show
```

```python
# Mudando a cor e legendando

plt.plot(empresas, bilhoes, color = 'green')
plt.title('Empresas Mais Ricas do Mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=15)
plt.show
```

```python
# Aumentando a espessura da linha lw

plt.plot(empresas, bilhoes, color = 'green', lw=10)
plt.title('Empresas Mais Ricas do Mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=15)
plt.show
```

```python
# Adicionando marcador

plt.plot(empresas, bilhoes, color = 'dimgray', marker = 'o')
plt.title('Empresas Mais Ricas do Mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=15)
plt.show
```

```python
# Adicionando linhas de grade

plt.plot(empresas, bilhoes, color = 'dimgray', marker = 'o')
plt.title('Empresas Mais Ricas do Mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
# Mudando o tipo do gráfico e colocando a legenda acima


plt.bar(empresas, bilhoes, color = 'dimgray')
plt.title('Empresas Mais Ricas do Mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = ' high right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
# Gráfico mais simples

plt.bar(empresas, bilhoes)
```

```python
cores = ['dimgray', 'dimgrey', 'gray', 'grey', 'darkgray', 'darkgrey', 'silver', 'lightgray', 'lightgrey', 'gainsboro']
```

```python
plt.bar(empresas, bilhoes, color = cores, edgecolor= 'black')
plt.title('Empresas mais ricas do mundo',  fontsize=25)
plt.ylabel('Bilhões de dólares', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
plt.grid(True)
```

```python
# Menu de estilos de gráficos
plt.style.available
```

```python
plt.style.use('dark_background')
plt.bar(empresas, bilhoes, color = cores, edgecolor= 'black')
plt.title('Empresas mais ricas do mundo',  fontsize=25)
plt.ylabel('Bilhões de dólares', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
#plt.grid(True)
```

```python
# barh - Mudando para barras horizontais

plt.style.use('dark_background')
plt.barh(empresas, bilhoes, color = cores, edgecolor= 'black')
plt.title('Empresas mais ricas do mundo',  fontsize=25)
plt.ylabel('Bilhões de dólares', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
#plt.grid(True)
```

```python
plt.style.use('dark_background')
plt.scatter(empresas, bilhoes, color = cores, edgecolor= 'black')
plt.title('Empresas mais ricas do mundo',  fontsize=25)
plt.ylabel('Bilhões de dólares', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
#plt.grid(True)
```
